### Converting numeric data types

```python
# Inspecting column values 
for col in ['actual_price','discounted_price','discount_percentage','rating','rating_count']:
    print(f"Unique sample values in {col}:")
    print(df[col].unique()[:10]) # show first 10 unique values
```

```python
# Remove commas from actual price and convert to float
df['actual_price'] = df['actual_price'].str.replace(',','').astype(float)
```

```python
# Remove commas from discounted price and convert to float
df['discounted_price'] = df['discounted_price'].str.replace(',','').astype(float)
```

```python
# Remove %symbol from discount_percentage and convert to flaot
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)
```

```python
print(df['rating'].unique())
```

```python
# Replace invalid '|' with NAN
df['rating'] = df['rating'].replace('|',np.nan)

# Convert safely to float
df['rating'] = pd.to_numeric(df['rating'], errors = 'coerce')

# Fill NAN with mean rating
df['rating']= df['rating'].fillna(df['rating'].mean())
```

```python
# Remove commas from rating_count and comvert to int
df['rating_count'] = df['rating_count'].str.replace(',','')
```

```python
# Convert safely to float
df['rating_count'] = pd.to_numeric(df['rating_count'], errors = 'coerce')
```

```python
# fill NaN with 0 (or median)
df['rating_count'] = df['rating_count'].fillna(0)
```

```python
# convert to int
df['rating_count'] = df['rating_count'].astype(int)
```

### Converting non numeric column to string and category type

```python
# Convert to string dtype
string_cols = ['product_id', 'product_name', 'user_id', 'user_name', 'review_title', 'review_content', 'about_product']
df[string_cols] = df[string_cols].astype("string")
```

```python
# Convert to category dtype
df['category'] = df['category'].astype("category")
```

```python
print(df.dtypes)
```

### Outlier detection and treatment

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols =  ['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']

for col in numeric_cols:
    plt.figure(figsize = (8,5))
    sns.boxplot(x=df[col], color = 'skyblue')
    plt.title(f"Boxplot of {col} (with outliers)")
    plt.xlabel(col)
    plt.show()
    
```

```python
df_clean = df.copy()
```

```python
def remove_outliers_iqr(df_clean, column):
    Q1 = df_clean[column].quantile(0.25)
    Q3 = df_clean[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df_clean[(df_clean[column] >= lower_bound) & (df_clean[column] <= upper_bound)]
num_cols = ['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']
for col in num_cols:
    df_clean = remove_outliers_iqr(df_clean, col)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in num_cols:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=df_clean[col], color="lightgreen")
    plt.title(f"Boxplot of {col} (after outlier treatment)")
    plt.xlabel(col)
    plt.show()
```

#### BUSINESS QUESTION:
2. How does the joint probability of ratings and promotional offers influence product comparison?

```python
# Define thresholds
high_rating_threshold = 4.0
high_discount_threshold = 50

# Boolean masks
high_rating = df_clean["rating"] >= high_rating_threshold
high_discount = df_clean["discount_percentage"] >= high_discount_threshold

# Joint condition
joint_condition = high_rating & high_discount

# Count of products satisfying the condition
joint_count = joint_condition.sum()

# Total number of products
total_count = len(df_clean)

# Joint probability
joint_probability = joint_count / total_count

print(f"Number of products with both high rating and high discount: {joint_count}")
print(f"Total number of products: {total_count}")
print(f"Joint probability: {joint_probability:.2%}")
```

INTERPRETATION :

 Around 40.83% pf products have both high rating and high discount. This indicates less than half of products combine strong user approval with aggressive promotions, which may present a valuable opportunity for targeted marketing.

#### BUSINESS QUESTION:
3. What is the conditional probability of selecting a product based on its performance compared to competitors?

If a product has a higher rating than its competitor, what is the probability it will be selected?

```python
# Retrieve their names
product_1 = top2_products[0]
product_2 = top2_products[1]

# Compute average ratings for each product
avg_rating_1 = df_clean.loc[df_clean["product_name"] == product_1, "rating"].mean()
avg_rating_2 = df_clean.loc[df_clean["product_name"] == product_2, "rating"].mean()

print(f"{product_1[:50]}... Average Rating: {avg_rating_1:.2f}")
print(f"{product_2[:50]}... Average Rating: {avg_rating_2:.2f}")

# Determine which has higher rating
if avg_rating_1 > avg_rating_2:
    higher_rated_product = product_1
elif avg_rating_2 > avg_rating_1:
    higher_rated_product = product_2
else:
    higher_rated_product = "Both (same rating)"

# Compute selection probability
total_reviews = df_clean[df_clean["product_name"].isin([product_1, product_2])].shape[0]
higher_rated_reviews = df_clean[df_clean["product_name"] == higher_rated_product].shape[0]

conditional_probability = higher_rated_reviews / total_reviews

print("\nConditional Probability:")
print(f"P(Selected | Higher Rating) = {conditional_probability:.2%}")
```

INTERPRETATION :

The product “MI Braided USB Type-C Cable” has a higher average rating (4.40) compared to “Wayona Nylon Braided USB Cable” (4.20).
Despite the better rating, only 50% of the total reviews for these two products belong to the higher-rated product.

#### BUSINESS QUESTION:
4. How can sampling techniques be employed to assess product preferences in the market?

```python
import numpy as np

# Fix the random seed for reproducibility
np.random.seed(42)

# Filter reviews of the two products you selected earlier
subset_df = df_clean[df_clean["product_name"].isin(top2_products)]

# Check how many rows are available
n_rows = len(subset_df)
print("Total records available:", n_rows)

# Pick a sample size (e.g., 50 or total rows if less)
sample_size = min(50, n_rows)

# Take a random sample without replacement
sample_df = subset_df.sample(n=sample_size, replace=False)

# Compute proportion of reviews per product in the sample
sample_counts = sample_df["product_name"].value_counts(normalize=True)

# Show results
print("\nSampled preference proportions:")
print(sample_counts)
```

INTERPRETATION :

Using random sampling of available reviews, both products appear equally preferred in the sample (50%-50%).
However, because the sample size is very small (only 6 reviews total), this proportion may not be representative of the broader customer base.

To get a more reliable estimate of preference:

You would need a larger dataset.

Or repeat this sampling many times (bootstrapping) to see the distribution of proportions.

#### SALES PERFORMANCE ANALYSIS

##### Do discounted products have significantly different sales volumes compared to non-discounted products?

HYPOTHESES

NULL HYPOTHESIS (H0) : Mean sales volume is equal for discounted and non-discounted products.
ALTERNATIVE HYPOTHESIS (H1) : Mean sales volume differs between discounted and non-discounted products.

Test Type : 
##### Two tailed sample t-test

```python
from scipy.stats import ttest_ind

discounted  = df[df["discount_percentage"] > 0]["rating_count"]
nondiscounted = df[df["discount_percentage"] == 0]["rating_count"]

t_stat, p_value = ttest_ind(discounted, nondiscounted, equal_var=False)

print("Two tailed t-test comparing mean sales volume:")
print("t-statistic:", t_stat)
print("p-value:",p_value)

if p_value < 0.05:
    print("Reject H0: There is significant difference in mean sales volume")
else :
    print("Fail to reject H0: No significant difference")
```

INTERPRETATION : 

since the p value is less than 0.05 we ewject the null hypothesis. 
This means there is strong evidence that the mean sales volume is significanlty different between discounted and non-discounted products.

##### Do products with higher discounts (>50%) have higher sales volumes than those with lower discounts (≤50%)?

HYPOTHESES 

NULL HYPOTHESIS (H0) : Mean sales volume for high discount <= mean sales volume for low discount.
ALTERNATIVE HYPOTHESIS (H1) : Mean sales volume for high discount > mean sales volume for low discount.

Test Type : 
##### One tailed sample t-test


```python
high_discount = df[df["discount_percentage"] > 50]["rating_count"]
low_discount = df[df["discount_percentage"] <= 50]["rating_count"]

t_stat, p_value_two_tailed = ttest_ind(high_discount, low_discount, equal_var = False)

p_value_one_tailed = p_value_two_tailed / 2

print("One-tailed t-test comparing high vs. low discount sales volume:")
print("t-statistic:", t_stat)
print("p-value (one-tailed):", p_value_one_tailed)

if (t_stat > 0) and (p_value_one_tailed < 0.05):
    print("Reject H0: High discount products have significantly higher sales volume.")
else:
    print("Fail to reject H0: No evidence that high discount products have higher sales volume.")
```

INTERPRETATION :

Since p value is greater than significane level 0.05 , we fail to reject the null hypothesis. 
This means there is no statistically significant evidence that products with higher discounts achieve higher average sales volumes compared to products with lower discounts.

##### two-tailed  t-test.

INTERPRETATION :
Since the p-value is far below the 0.05 significance level, we reject the null hypothesis.

#### IMPROVING THE MODEL BY ADDING FEATURES AND CHANGING VALUE OF K 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
features = df_clean[["discount_percentage","discounted_price","rating","actual_price","rating_count"]]

df_clean["high_sales"] = (df_clean["rating_count"] >= df_clean["rating_count"].median()).astype(int)

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

X_train, X_test, y_train, y_test = train_test_split(scaled_features, df_clean["high_sales"], test_size=0.2, random_state=42)

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy: ", accuracy)
print("Confusion matrix:\n", conf_matrix)
print("Classification report:\n", class_report)
```

INTERPRETATION :

  This is a strong, balanced model that can confidently classify high and low sales products.
  We improved the KNN model by including additional predictors (rating, actual_price) and optimizing the number of neighbors (k=7). This led to an accuracy increase from ~58% to ~93%. The model demonstrates high precision and recall for both classes, confirming that richer features significantly enhance predictive power.
